```
# Tensorflow (3.8 tensorflow, 3.7: ts 2.0.0)
# import keras -> tensorflow.keras
from tensorflow.keras.datasets.mnist import load_data
# ((x_train, y_train), (x_test, y_test))
print(load_data())
(x_train, y_train), (x_test, y_test) = load_data()
```

```
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```
import matplotlib.pyplot as plt
# matplotlib.pyplot.imshow(X, cmap=None(顏色), norm=None, aspect=None, interpolation=None, alpha=None, vmin=None, vmax=None, origin=None, extent=None, *, filternorm=True, filterrad=4.0, resample=None, url=None, data=None, **kwargs)[source]
# _r reverse color
plt.imshow(x_train[0], cmap= "gray_r")
```

```
import pandas as pd
pd.DataFrame(x_train[0])
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

layers = [
          # 784 * 128/256 + 128/256 bias (前置偏移調整)
          Dense(256, activation = "relu", input_dim = 784),
          # 128 * 10 + 10 bias (前置偏移調整)
          Dense(10, activation = "softmax")
]

model = Sequential(layers)
model.summary()
```

```
# 一個神經元（兩個機率）: BinaryCrossEntropy
# 多個神經元: pilog(1/qi): CategoricalCrossEntropy
# model.compile(loss= "mse", 不用mse 效果差
# MSE Loss效果:  1/n (from i to n(PREi - ACCi)**2 ) v.s. one-hot encoding 轉為某項機率為1 其餘為零
from tensorflow.keras.losses import CategoricalCrossentropy
model.compile(loss= CategoricalCrossentropy() ,
              metrics = ["accuracy"],
              optimizer = "adam")
```

```
from tensorflow.keras.utils import to_categorical
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)
print("Before one-hot:",y_train[0])
print("After one-hot: ",y_train_cat[0])
```

```
# reshape可以填一個(-1)讓電腦算 x_train.reshape(60000, 784)＝(-1, 784)
x_train_shape= x_train.reshape(-1, 784) / 255.0
x_test_shape = x_test.reshape(-1, 784).astype("float")/255
  
```

```
# batch_size : 調整方向（平均去調避免走歪路）->圖片大/data多:少載batch 圖片小/data 少: 多載batch (0-200 OK)
# prevent overfitting: validation_split(特殊資料不學)+ 看到epochs 重看val_loss幾次穩定就停
# epochs 調整重看次數, verbose=進度條
# callback (ex: 綠界 付款完再回去呼叫原本網站 -> 每輪epoch 就 callback): EarlyStopping(monitor = val_loss, patience= 幾次沒變好就停, restore_best_weight= 回朔找到最好的)
from tensorflow.keras.callbacks import EarlyStopping
callbacks = [
             EarlyStopping(patience = 5, restore_best_weights = True)
]
model.fit(x_train_shape,
          y_train_cat,
          batch_size = 200,
          epochs = 50,
          verbose = 2,
          validation_split = 0.1,
          callbacks= callbacks)
```

```
# evaluate完去調dense 128 -> 256/64/512去找到最佳
model.evaluate(x_test_shape, y_test_cat)
```

```
from sklearn.metrics import confusion_matrix
import pandas as pd

# sklearn: predict_proba/predict
# tensorflow: predict/predict_classes (代y_test_cat), 不是tensorflow 代y_test
# confusion_matrix(y_true, y_pre)

pre = model.predict_classes(x_test_shape)
mat = confusion_matrix(y_test, pre)
pd.DataFrame(mat,
             columns = [str(i) + " Predict" for i in range(10)],
             index = [str(i) + " Accuracy" for i in range(10)])
```

```
# subplot(高, 寬, ith) 從一開始數
import numpy as np
# no.nonzero 找非零值
# np.nonzero(pre != y_test)[0]
print(np.nonzero(pre != y_test)[0])
idx = np.nonzero(pre != y_test)[0]
false_img = x_test[idx]
false_pre = pre[idx]
false_label = y_test[idx]
plt.figure(figsize=(14,42))
width = 10
height = len(idx)// width + 1
for i in range(len(idx)):
  plt.subplot(height, width, i +1)
  plt.imshow(false_img[i])
  t = "[O]:{}\n[P]:{}".format(false_label[i], false_pre[i])
  plt.title(t)
  plt.axis("off")
```
